# Chapter 1-2. Bayesian modeling fundamentals

```python
%matplotlib inline

import numpy as np
import pandas as pd
from scipy import stats
import seaborn as sns
import matplotlib.pyplot as plt
import arviz as az
```

```python
N = 10  # Number of samples
FIGSIZE = (16, 9)
```

## Define simple model with probabilistic distribution for linear function

As a first step, we will start to model linear function based on normal distributions as follows:  

$$
\begin{equation*}
\begin{aligned}

y = ax + b  \\
a \sim \mathcal{N(0.0, 1.0)} \\
b \sim \mathcal{N(0.0, 1.0)}

\end{aligned}
\end{equation*}
$$

```python
x = np.linspace(-3, 3, 100)
plt.figure(figsize=FIGSIZE)

for i in range(N):
    a = stats.norm(0, 1).rvs()
    b = stats.norm(0, 1).rvs()
    plt.plot(x, a * x + b, label=f'$a={np.round(a, 2)} / b={np.round(b, 2)}')

plt.legend(loc='upper center', bbox_to_anchor=(0.5, 1.05),
    ncol=3, fancybox=True, shadow=True)
plt.show()
```

Then what if one of the paramter is generated from another type of distribution, i.e. Gamma distribution?  

$$
\begin{equation*}
\begin{aligned}

y = ax + b \\
a \sim \mathcal{Gamma(1.0, 1.0)} \\
b \sim \mathcal{N(0.0, 1.0)}

\end{aligned}
\end{equation*}
$$

```python
x = np.linspace(-3, 3, 100)
plt.figure(figsize=FIGSIZE)

for i in range(N):
    a = stats.gamma(1.0, 1.0).rvs()
    b = stats.norm(0, 1.).rvs()
    plt.plot(x, a * x + b, label=f'$a={np.round(a, 2)} / b={np.round(b, 2)}')

plt.legend(loc='upper center', bbox_to_anchor=(0.5, 1.05),
    ncol=3, fancybox=True, shadow=True)
plt.show()
```

## Define generalized linear model

Finally, we will look at the generalized linear model defined as follows:  

$$
\begin{equation}
\begin{aligned}

y = exp(ax + b) \\
a \sim \mathcal{N(0.0, 1.0)}  \\
b \sim \mathcal{N(0.0, 1.0)}

\end{aligned}
\end{equation}
$$

```python
x = np.linspace(-3, 3, 100)
plt.figure(figsize=FIGSIZE)

for i in range(N):
    a = stats.norm(0, 1.).rvs()
    b = stats.norm(0, 1.).rvs()
    plt.plot(x, np.exp(a * x + b), label=f'$a={np.round(a, 2)} / b={np.round(b, 2)}')

plt.legend(loc='upper center', bbox_to_anchor=(0.5, 1.05),
    ncol=3, fancybox=True, shadow=True)
plt.show()
```
